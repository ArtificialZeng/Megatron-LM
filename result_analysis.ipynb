```python
gen_true_path='/gpfs/fs1/projects/gpu_adlr/datasets/dasu/prompting/predicted/TQA/GenCTX/generated_context_k10_357m_multisetdpr_queryctx_p0.9_all.true_list.txt'
gen_false_path='/gpfs/fs1/projects/gpu_adlr/datasets/dasu/prompting/predicted/TQA/GenCTX/generated_context_k10_357m_multisetdpr_queryctx_p0.9_all.false_list.txt'
ret_true_path='/gpfs/fs1/projects/gpu_adlr/datasets/dasu/prompting/predicted/TQA/GenCTX/generated_context_k10_357m_multisetdpr_queryctx_p0.9_all.retrieval.true_list.txt'
ret_false_path='/gpfs/fs1/projects/gpu_adlr/datasets/dasu/prompting/predicted/TQA/GenCTX/generated_context_k10_357m_multisetdpr_queryctx_p0.9_all.retrieval.false_list.txt'
```

```python
gen_true_list, gen_false_list, ret_true_list, ret_false_list = [], [], [], []

with open(gen_true_path, 'r') as f:
    for each in f.readlines():
        gen_true_list.append(each.strip('\n'))
with open(gen_false_path, 'r') as f:
    for each in f.readlines():
        gen_false_list.append(each.strip('\n'))

with open(ret_true_path, 'r') as f:
    for each in f.readlines():
        ret_true_list.append(each.strip('\n'))

with open(ret_false_path, 'r') as f:
    for each in f.readlines():
        ret_false_list.append(each.strip('\n'))
```

```python
rt_gt_list = []
rt_gf_list = []
rf_gt_list = []
rf_gf_list = []

for each in gen_true_list:
    if each in ret_true_list:
        rt_gt_list.append(each)
    else:
        rf_gt_list.append(each)

for each in gen_false_list:
    if each in ret_true_list:
        rt_gf_list.append(each)
    else:
        rf_gf_list.append(each)


```
